```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.zeros((2000, 2000))
pixsize = 0.1
NA_outer = 0.38
NA_inner = 0.22
index = 1.33
```

```python
half_shape = array.shape[0]//2
# # -half to half
x = np.arange(-1, 1+1/half_shape, 1/half_shape)/(2*pixsize)
print(x)
```

```python
y = x
[X, Y] = np.meshgrid(x, y)
R = np.sqrt(X * X + Y * Y)
MaxRad = NA_outer / index  # maximum annulus diameter
MinRad = NA_inner / index  # minimum annulus diameter
AnnularFilter = (R <= MaxRad) & (R >= MinRad)
```

```python
plt.imshow(AnnularFilter)
```
